```python
# Day 1: Data Exploration Notebook
# File: notebooks/01_data_exploration.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
url = "https://raw.githubusercontent.com/sairaawahid/Employee-Attrition-Predictor-for-HR-Analytics/refs/heads/main/data/WA_Fn-UseC_-HR-Employee-Attrition.csv?token=GHSAT0AAAAAADFEKMZXWRLB6NQDWT2SN4I22CBS6GQ"
df = pd.read_csv(url)

# Quick Glance
print("âœ… Dataset loaded successfully!\n")
print("ðŸ”Ž First 5 rows:")
display(df.head())

print("\nðŸ§¾ Dataset Info:")
df.info()

print("\nðŸ“Š Statistical Summary:")
display(df.describe(include='all'))

# Check for nulls and duplicates
print(f"\nðŸ§¼ Missing values:\n{df.isnull().sum()}")
print(f"\nðŸ§¼ Duplicates: {df.duplicated().sum()}")

# Basic Visualization - Attrition Count
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Attrition')
plt.title("Employee Attrition Distribution")
plt.show()

# Satisfaction vs Attrition
plt.figure(figsize=(8, 5))
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df)
plt.title("Job Satisfaction by Attrition")
plt.show()

# Age vs Attrition
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', hue='Attrition', kde=True, multiple='stack')
plt.title("Age Distribution by Attrition")
plt.show()
```
